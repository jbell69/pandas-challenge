```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

purchase_data_df.head(20)
```

## Player Count

* Display the total number of players


```python
players = {'Total Players':[len(purchase_data_df['SN'].value_counts())]}

total_players = pd.DataFrame(players)
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data_df['Item ID'].nunique()
Average_price = purchase_data_df["Price"].mean()
total_purchases = purchase_data_df['Item Name'].count()
Total_revenue = purchase_data_df["Price"].sum()
dollars = "${0:.2f}".format
percentage = "{0:.2f}%".format


Display_Summary_data = [{"Number of Unique Items":unique_items, "Average Price":Average_price,
                         "Total Purchases":total_purchases,"Total Revenues":Total_revenue}]


Display_Summary_table = pd.DataFrame(Display_Summary_data)
Display_Summary_table["Average Price"] = Display_Summary_table["Average Price"].map(dollars)
Display_Summary_table["Total Revenues"] = Display_Summary_table["Total Revenues"].map(dollars)

Display_Summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
grouped_gender_df = purchase_data_df.groupby(['Gender'])
total_unique_gender = grouped_gender_df['SN'].nunique()

gender_percentage = purchase_data_df

percentage_players = (total_unique_gender/[len(purchase_data_df['SN'].value_counts())])*100


summary_df = pd.DataFrame({"Total Count" : total_unique_gender, "Percentage of Players" : percentage_players })

summary_df["Percentage of Players"] = summary_df["Percentage of Players"].map(percentage)

summary_df
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

Group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age groups"] = pd.cut(purchase_data_df["Age"], bins, labels=Group_names)
purchase_data_df_age = purchase_data_df.groupby("Age groups")

total_purchases_age = purchase_data_df_age['SN'].nunique()

total_purchases_age_percent = (purchase_data_df_age['SN'].nunique() / [len(purchase_data_df['SN'].value_counts())])*100

summary_df_age = pd.DataFrame({"Total Count" : total_purchases_age, 
                               "Percentage of Players" : total_purchases_age_percent})

summary_df_age["Percentage of Players"] = summary_df_age["Percentage of Players"].map(percentage)

summary_df_age
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

Group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age groups"] = pd.cut(purchase_data_df["Age"], bins, labels=Group_names)
purchase_data_df_age = purchase_data_df.groupby("Age groups")

total_purchases_age_count = purchase_data_df_age['Age'].size()
#total_purchases_age_count

average_purchases_age_count = purchase_data_df_age['Price'].mean().round(2)
#average_purchases_age_count

total_revenue_age_count = purchase_data_df_age['Price'].sum()
#total_revenue_age_count

Total_revenue_per_player_age_count = (total_revenue_age_count/[len(purchase_data_df_age['SN'].value_counts())])*100
#Total_revenue_per_player_age_count

summary_df_age_count = pd.DataFrame({"Purchase Count" : total_purchases_age_count, 
                               "Average Purchase Price" : average_purchases_age_count, 
                               "Total Purchase Value" : total_revenue_age_count, 
                               "Avg Total Purchase per Person" : Total_revenue_per_player_age_count})

summary_df_age_count["Average Purchase Price"] = summary_df_age_count["Average Purchase Price"].map(dollars)
summary_df_age_count["Total Purchase Value"] = summary_df_age_count["Total Purchase Value"].map(dollars)
summary_df_age_count["Avg Total Purchase per Person"] = summary_df_age_count["Avg Total Purchase per Person"].map(dollars)

summary_df_age_count
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

total_spender = purchase_data_df.groupby('SN')

total_spender_count = total_spender['Purchase ID'].count()

total_spender_average = total_spender['Price'].mean()

total_spender_revenue = total_spender['Price'].sum()


summary_df_spender = pd.DataFrame({"Purchase Count" : total_spender_count, 
                               "Average Purchase Price" : total_spender_average, 
                               "Total Purchase Value" : total_spender_revenue})

summary_df_table = summary_df_spender.sort_values('Total Purchase Value', ascending=False)

summary_df_table["Average Purchase Price"] = summary_df_table["Average Purchase Price"].map(dollars)
summary_df_table["Total Purchase Value"] = summary_df_table["Total Purchase Value"].map(dollars)

summary_df_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order

```python
most_popular = purchase_data_df.groupby(['Item ID', 'Item Name'])
most_popular.count()
most_popular_count = most_popular['Purchase ID'].count()

most_popular_average = most_popular['Price'].mean()

most_popular_revenue = most_popular['Price'].sum()


most_popular_spender = pd.DataFrame({"Purchase Count" : most_popular_count, 
                                     "Average Purchase Price" : most_popular_average, 
                                     "Total Purchase Value" : most_popular_revenue})

most_popular_table = most_popular_spender.sort_values('Purchase Count', ascending=False)

most_popular_table["Average Purchase Price"] = most_popular_table["Average Purchase Price"].map(dollars)
most_popular_table["Total Purchase Value"] = most_popular_table["Total Purchase Value"].map(dollars)

most_popular_table.head(20)
```
